## Übung 3: Variablen über Code-Blöcke hinweg

```python
# In Jupyter Notebooks bleiben Variablen gespeichert, auch wenn sie in verschiedenen Code-Blöcken stehen.
# Das bedeutet: Einmal definierte Variablen kannst du in allen nachfolgenden Blöcken verwenden!

# Führe zuerst diesen Code-Block aus
lieblingsessen = "Pizza"
lieblingszahl = 10
```

```python
# Jetzt führe diesen Code-Block aus (ohne den oberen Block nochmal auszuführen)
# Beobachte: Die Variablen aus dem ersten Block sind noch verfügbar!
print("Mein Lieblingsessen ist", lieblingsessen)
print("Meine Lieblingszahl ist", lieblingszahl)
print("Das Doppelte meiner Lieblingszahl ist", lieblingszahl * 2)

# Aufgabe: Gehe zurück zum ersten Block, ändere die Werte und führe dann beide Blöcke erneut aus
```
