```python
def berechne_paritaetsbit(bitfolge):
    """
    Berechnet das Paritätsbit für eine gegebene Bitfolge.
    
    Parameter:
    bitfolge (str): String aus '0' und '1'
    parität (str): 'gerade' oder 'ungerade' (Standard: 'gerade')

    prints: bitfolge mit parität am Ende
    """

    count = 0
    for i in bitfolge:
        if i == "1":
            count = count + 1
    if count % 2 == 0:
        print(bitfolge + "0")
    else:
        print(bitfolge + "1")

berechne_paritaetsbit("10101111")
```

## Was ist ein Paritätsbit?

Ein Paritätsbit ist ein zusätzliches Bit, das an eine Bitfolge angehängt wird, um einfache Übertragungsfehler zu erkennen.
